#Higher fidelity data with zipcodes

```python
newOrleansDFs = {}

#Load in older dataset for 2015-2016 data
csv_files = glob.glob("/content/drive/MyDrive/Work/Citadel 2021/Data/New Orleans" + "/*.csv")
#Create regex pattern to identify date of each listing
regexPattern = re.compile("_(201.+?(?=\.csv))")

#load each of the New Orleans airbnb files, with the length of each csv being the # of listings in the month
for filename in csv_files:
  newOrleansDFs[pd.to_datetime(regexPattern.findall(filename)[0]).strftime("%B %Y")] = pd.read_csv(filename)

#len(newOrleansAirbnb.keys()) == len(csv_files)

#Load in Professor dataset for 2017-2019 data
csv_files = glob.glob("/content/drive/MyDrive/Work/Citadel 2021/Data/New Orleans Professor" + "/*.csv")
#Create regex pattern to identify date of each listing

#load each of the New Orleans airbnb files, with the length of each csv being the # of listings in the month
for filename in csv_files:
  newOrleansDFs[pd.to_datetime("20" + filename[-6:-4] + "-" + filename[-8:-6] + "-"+ "01").strftime("%B %Y")] = pd.read_csv(filename)
  
regexPattern = re.compile("([^\/]+$)")
csv_files = glob.glob("/content/drive/MyDrive/Work/Citadel 2021/Data/New Orleans 2020-2021" + "/*.csv")
for filename in csv_files:
  newOrleansDFs[regexPattern.findall(filename)[0][:-4]] = pd.read_csv(filename)

columnsOfInterest = ["date", "price", "neighborhood", "latitude", "longitude", "reviews"]

newOrleansListings = pd.DataFrame(columns = columnsOfInterest)

for j in newOrleansDFs.keys():
  if "neighbourhood" in newOrleansDFs[j].columns:
    newOrleansDFs[j].rename(columns = {"neighbourhood": "neighborhood"}, inplace = True)
  if "number_of_reviews" in newOrleansDFs[j].columns:
    newOrleansDFs[j].rename(columns = {"number_of_reviews": "reviews"}, inplace = True)
  if "weekly_price" in newOrleansDFs[j].columns and "price" not in newOrleansDFs[j].columns:
    newOrleansDFs[j]["price"] = np.nan
  if "neighbourhood_cleansed" in newOrleansDFs[j].columns:
    newOrleansDFs[j].drop(columns = ["neighborhood"])
    newOrleansDFs[j]["neighborhood"] = newOrleansDFs[j]["neighbourhood_cleansed"]

  newOrleansDFs[j]["neighborhood"] = newOrleansDFs[j]["neighborhood"].str.lower()
  newOrleansDFs[j]["date"] = [j]*len(newOrleansDFs[j])
  newOrleansListings = pd.concat([newOrleansListings, newOrleansDFs[j].loc[:,columnsOfInterest]], join = "outer", ignore_index = True)

newOrleansListings
```
